```python
import os, re, requests, zipfile

# === Paste your text below ===
data = """
[Lecture] ELECTRICIAN PDF: https://appx-content-v2.classx.co.in/paid_course4/2025-01-13-0.8211423884710038.pdf@KEY=@TYPE=0.00
[Lecture] Intro Class Er. S K Jha sir: https://www.masterapi.tech/get/appx/signedUrl/354737.598.0.811632.zip
[Basic of Electricity] Basic of Electricity Class-1: https://www.masterapi.tech/get/appx/signedUrl/375756.598.0.811632.zip
...
"""  # You can paste your full list here

os.makedirs("downloads", exist_ok=True)
lines = [line.strip() for line in data.strip().splitlines() if line.strip()]

def safe_filename(name):
    return re.sub(r'[<>:"/\\|?*]', '_', name).strip()

def download_file(url, filename):
    print(f"â¬‡ï¸ Downloading: {filename}")
    r = requests.get(url, stream=True)
    if r.status_code != 200:
        print(f"âŒ Failed ({r.status_code}) â†’ {url}")
        return None
    with open(filename, "wb") as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)
    print(f"âœ… Done: {filename}")
    return filename

def extract_zip(zip_path, target_dir):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(target_dir)
    print(f"ğŸ“¦ Extracted to: {target_dir}")

for line in lines:
    if ":" not in line:
        continue
    name, url = line.split(":", 1)
    name = safe_filename(name.strip().replace("[", "").replace("]", ""))
    url = url.strip()
    folder = os.path.join("downloads", name)
    os.makedirs(folder, exist_ok=True)

    if ".zip" in url:
        zip_name = os.path.join(folder, name + ".zip")
        path = download_file(url, zip_name)
        if path:
            try:
                extract_zip(path, folder)
                os.remove(path)
            except zipfile.BadZipFile:
                print(f"âš ï¸ Invalid ZIP: {path}")
    elif ".pdf" in url:
        pdf_name = os.path.join(folder, name + ".pdf")
        download_file(url, pdf_name)

print("\nğŸ¯ All downloads complete! Check the 'downloads' folder.")
```
